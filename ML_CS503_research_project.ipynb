#**Task 4**

For calculation of MCE, We have used the procedure given in reference paper[1]<br>

Procedure : we changed the main py function to return(save as .npy file) test_data and predicted_probability for test data.<br>
image.png

Now, we use these values mainly (test_data and preddiced probability) to calculate MCE error.<br>
Code for MCE error is given in below cell.

We first divide the dataset into 'num_bins' equal (nearly) parts randomly.<br>
In each bin, we calculate the confidence (the probability associated with the predicted class) for each sample.<br>
Now we define the accuracy of a bin as the usual accuracy (total_correct_pred/total_pred).<br>
And we define the confidence of a bin by the mean of the confidence of all samples in the bin.<br>
<br>
Now the caliberation error of a bin is absolute difference of accuracy and confidence.<br>
<br>
We calculated the caliberation error of all bins and took **maximum** of them and reported as **MCE**
<br>
**Reference**

image.png

For platt scaling, we have used **Temperature Platt Scaling**<br>
image.png

To calculate this, we applied logit (sigmoid inverse) to the best (due to maximization in definition) <br>
confidence and divided that by 'T_temp_plat' and again alpplied sigmoid <br>function to get new confidence as per the definition above.

**NOTE:** instead of running the model again and calculating the MCE after applying platt scaling,<br> we can calculate it using old predicted probabilities only.<br>Please see the MCE function implemented below. <br>It can calculate MCE and also it has a hyperparameter 'platt_scaling' which<br> if set to true will apply Temperature platt scaling and then calculate MCE.

### Function to calculate MCE and platt scaling

```python
import random
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

def logit(p):
    return p/(1-p)

def MCE(test_data,pred_prob,num_bins,platt_scaling=False,T_temp_plat = 5):
    y_pred_test_all = np.array([])
    y_true_test_all = np.array([])
    num_devices = len(test_data)
    confidence = np.array([])

    for i in range(num_devices):
        y_true_cur_device = test_data[i]['y']
        y_true_test_all = np.concatenate((y_true_test_all,y_true_cur_device),axis=0)
        y_pred_cur_device = np.argmax(pred_prob[i],axis=1)
        y_pred_test_all = np.concatenate((y_pred_test_all,y_pred_cur_device),axis=0)
        confidence_cur_device = np.max(pred_prob[i],axis=1)

        if platt_scaling==True:
            z = logit(confidence_cur_device)
            confidence_cur_device = sigmoid(z/T_temp_plat)
        confidence = np.concatenate((confidence,confidence_cur_device),axis=0)

    num_tot_samples = len(confidence)
    index = [i for i in range(num_tot_samples)]
    bin_size = num_tot_samples//num_bins
    bins_list = []
    random.seed(10)
    for i in range(num_bins):
        cur_bin = random.sample(index,bin_size)
        for x in cur_bin:
            index.remove(x)
        bins_list.append(np.array(cur_bin))

    bins_caliberation_error = []
    for i in range(num_bins):
        cur_bin_index = bins_list[i]
        cur_y_true_test = y_true_test_all[cur_bin_index]
        cur_y_pred_test = y_pred_test_all[cur_bin_index]

        cur_bin_accuracy = np.mean(cur_y_true_test == cur_y_pred_test)
        cur_bin_confidence = np.mean(confidence[cur_bin_index])

        current_bin_caliberation_error = abs(cur_bin_accuracy-cur_bin_confidence)
        bins_caliberation_error.append(current_bin_caliberation_error)

    MCE = np.max(bins_caliberation_error)
    return MCE
```
